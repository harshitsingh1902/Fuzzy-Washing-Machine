```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
dirtness = ctrl.Antecedent(np.arange(0, 101, 1), 'dirtness')
typeofdirt = ctrl.Antecedent(np.arange(0, 101, 1), 'typeofdirt')
typeoffabric = ctrl.Antecedent(np.arange(0, 101, 1), 'typeoffabric')
volumeofclothes = ctrl.Antecedent(np.arange(0, 101, 1), 'volumeofclothes')

washingtime = ctrl.Consequent(np.arange(0, 101, 1), 'washingtime')
washingspeed = ctrl.Consequent(np.arange(0, 1201, 1), 'washingspeed')
waterintake = ctrl.Consequent(np.arange(0, 101, 1), 'waterintake')
watertemp = ctrl.Consequent(np.arange(0, 81, 1), 'watertemp')
```

```python
dirtness['small'] = fuzz.trimf(dirtness.universe, [0, 0, 50])
dirtness['medium'] = fuzz.trimf(dirtness.universe, [10, 50, 90])
dirtness['large'] = fuzz.trimf(dirtness.universe, [50, 100, 100])

typeofdirt['notgreasy'] = fuzz.trimf(typeofdirt.universe, [0, 0, 50])
typeofdirt['medium'] = fuzz.trimf(typeofdirt.universe, [10, 50, 90])
typeofdirt['greasy'] = fuzz.trimf(typeofdirt.universe, [50, 100, 100])

typeoffabric['silk'] = fuzz.trimf(typeoffabric.universe, [0, 0, 50])
typeoffabric['woolen'] = fuzz.trimf(typeoffabric.universe, [10, 50, 90])
typeoffabric['cotton'] = fuzz.trimf(typeoffabric.universe, [50, 100, 100])

volumeofclothes['small'] = fuzz.trimf(volumeofclothes.universe, [0, 0, 50])
volumeofclothes['medium'] = fuzz.trimf(volumeofclothes.universe, [10, 50, 90])
volumeofclothes['large'] = fuzz.trimf(volumeofclothes.universe, [50, 100, 100])

washingtime['veryshort'] = fuzz.trimf(washingtime.universe, [0, 0, 20])
washingtime['short'] = fuzz.trimf(washingtime.universe, [20, 30, 40])
washingtime['medium'] = fuzz.trimf(washingtime.universe, [30, 50, 70])
washingtime['long'] = fuzz.trimf(washingtime.universe, [50, 70, 90])
washingtime['verylong'] = fuzz.trimf(washingtime.universe, [80, 100, 100])

washingspeed['verylow'] = fuzz.trimf(washingspeed.universe, [0, 0, 450])
washingspeed['slow'] = fuzz.trimf(washingspeed.universe, [350, 500, 700])
washingspeed['medium'] = fuzz.trimf(washingspeed.universe, [550, 700, 900])
washingspeed['fast'] = fuzz.trimf(washingspeed.universe, [700, 900, 1000])
washingspeed['veryfast'] = fuzz.trimf(washingspeed.universe, [900, 1200, 1200])

waterintake['little'] = fuzz.trimf(waterintake.universe, [0, 0, 50])
waterintake['normal'] = fuzz.trimf(waterintake.universe, [10, 50, 90])
waterintake['alotof'] = fuzz.trimf(waterintake.universe, [50, 100, 100])

watertemp['low'] = fuzz.trimf(watertemp.universe, [0, 0, 50])
watertemp['normal'] = fuzz.trimf(watertemp.universe, [10, 50, 60])
watertemp['high'] = fuzz.trimf(watertemp.universe, [50, 80, 80])
```

```python
rule1 = ctrl.Rule(dirtness['medium'] & typeofdirt['medium'] & typeoffabric['silk'] & volumeofclothes['small']
                  , (washingtime['veryshort'], washingspeed['slow'], waterintake['normal'], watertemp['low']))
rule2 = ctrl.Rule(dirtness['medium'] & typeofdirt['medium'] & typeoffabric['silk'] & volumeofclothes['medium']
                  , (washingtime['short'], washingspeed['fast'], waterintake['normal'], watertemp['low']))
rule3 = ctrl.Rule(dirtness['medium'] & typeofdirt['medium'] & typeoffabric['woolen'] & volumeofclothes['small']
                  , (washingtime['veryshort'], washingspeed['slow'], waterintake['normal'], watertemp['low']))
rule4 = ctrl.Rule(dirtness['medium'] & typeofdirt['medium'] & typeoffabric['woolen'] & volumeofclothes['medium']
                  , (washingtime['short'], washingspeed['fast'], waterintake['normal'], watertemp['low']))
rule5 = ctrl.Rule(dirtness['medium'] & typeofdirt['notgreasy'] & typeoffabric['silk'] & volumeofclothes['small']
                  , (washingtime['medium'], washingspeed['verylow'], waterintake['normal'], watertemp['low']))
rule6 = ctrl.Rule(dirtness['medium'] & typeofdirt['notgreasy'] & typeoffabric['silk'] & volumeofclothes['medium']
                  , (washingtime['medium'], washingspeed['medium'], waterintake['normal'], watertemp['low']))
rule7 = ctrl.Rule(dirtness['medium'] & typeofdirt['notgreasy'] & typeoffabric['woolen'] & volumeofclothes['small']
                  , (washingtime['medium'], washingspeed['medium'], waterintake['normal'], watertemp['low']))
rule8 = ctrl.Rule(dirtness['medium'] & typeofdirt['notgreasy'] & typeoffabric['woolen'] & volumeofclothes['medium']
                  , (washingtime['medium'], washingspeed['medium'], waterintake['normal'], watertemp['low']))

# similarly other (81 - 8) rules can be defined. (other rules are not used in our case)
```

```python
rule1.view
rule1.view()
```

```python
washingmachine_ctrl = ctrl.ControlSystem(
    [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8])
```

```python
washingmachine = ctrl.ControlSystemSimulation(washingmachine_ctrl)
```

```python
washingmachine.input['dirtness'] = 50
washingmachine.input['typeofdirt'] = 20.8
washingmachine.input['typeoffabric'] = 21.5
washingmachine.input['volumeofclothes'] = 28
```

```python
washingmachine.compute()
```

```python
print(washingmachine.output['washingtime'])
print(washingmachine.output['washingspeed'])
print(washingmachine.output['waterintake'])
print(washingmachine.output['watertemp'])
```
